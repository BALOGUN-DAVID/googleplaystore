##Exploring App Categories

```python
#checking for the total number of unique categories
num_categories = df['Category'].nunique()
print('Number of categories = ', num_categories)
```

```python
#Checking for all the unique apps in the category column
df['Category'].unique()
```

```python
#'1.9' was seen as an app in the category column which shows irregularity and accuracy problem. Checking the exact position of this row
df[df['Category']== '1.9']
```

```python
#'1.9' is found on row 10472. Using numpy to convert it to NAN so as to get rid of it from the dataset since it is not up to 1% of the data, it definitely won't be having a significant effect on the analysis/insight gotten from this exploration
df.loc[10472]=df.loc[10472].shift()
df['App'].loc[10472]=df['Category'].loc[10472]
df['Category'].loc[10472]=np.nan
df.loc[10472]
```

```python
#Counting the number of apps in each category
count=df['Category'].value_counts()
count
```

```python
#converting the value count to a dataframe, in order to create a column for the counts
Count=pd.DataFrame(count)
Count.columns=['count1']
Count.head()
```

```python
#creating a new dataframe from the count of category column we worked on earlier
num_count =pd.DataFrame(df['Category'])
#value_counts()
num_count
```

```python
#combined the two dataframes now and grouped it by the category column while also resetting the indexes using the reset_index() method
sorted_category = num_count.groupby(['Category']).size().reset_index(name='Count')
sorted_category.head(10)
```

```python
#sorting the dataframe by the Count column in descending order
num_apps_in_category = pd.DataFrame(sorted_category.sort_values(by=['Count'], ascending=False))
num_apps_in_category.head(10)
```

```python
#Importing the necessary libraries needed to visualize the Apps in each category while showing the market prevalence
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=num_apps_in_category['Category'], y=num_apps_in_category['Count'])
plt.xticks(rotation=80)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Market Prevalence')
plt.show()
```

**Size and price of an app**

```python
import numpy as np
df['Size'].unique()
chars_to_remove =['M']
cols_to_clean = ['Size']
for col in cols_to_clean:
    for chars in chars_to_remove:
        df[col] = df[col].apply(lambda x:x.replace(chars, '000'))
        chars_to_remove =['.']
cols_to_clean = ['Size']
for col in cols_to_clean:
    for chars in chars_to_remove:
        df[col] = df[col].apply(lambda x:x.replace(chars, ''))
df['Size'].unique()
chars_to_remove =['k']
cols_to_clean = ['Size']
for col in cols_to_clean:
    for chars in chars_to_remove:
        df[col] = df[col].apply(lambda x:x.replace(chars, ''))
df['Size'].unique()
chars_to_remove =['Varies with device']
cols_to_clean = ['Size']
for col in cols_to_clean:
    for chars in chars_to_remove:
        df[col] = df[col].apply(lambda x:x.replace(chars, ('0')))
df['Size'].unique()
```

```python
import numpy as np
df['Size'] = df['Size'].astype('float')
df
```

```python
df.dtypes
```

```python
%matplotlib inline
import seaborn as sns
sns.set_style("darkgrid")
import warnings
warnings.filterwarnings("ignore")

# Select rows where both 'Rating' and 'Size' values are present (ie. the two values are not null)
apps_with_size_and_rating_present = df[~df[['Rating', 'Size']].isnull().any(axis = 1)]

# Subset for categories with at least 250 apps
large_categories = apps_with_size_and_rating_present.groupby(['Category']).filter(lambda x: len(x) >= 250)

# Plot size vs. rating
plt1 = sns.jointplot(x = large_categories['Size'], y = large_categories['Rating'])

# Select apps whose 'Type' is 'Paid'
paid_apps = apps_with_size_and_rating_present[apps_with_size_and_rating_present['Type'] == 'Paid']

# Plot price vs. rating
plt2 = sns.jointplot(x = paid_apps['Price'], y = paid_apps['Rating'])
plt.show()
```

**Relation between app category and app price**

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
fig.set_size_inches(15, 8)
import seaborn as sns

# Select a few popular app categories
popular_app_cats = df[df.Category.isin(['GAME', 'FAMILY', 'PHOTOGRAPHY',
                                            'MEDICAL', 'TOOLS', 'FINANCE',
                                            'LIFESTYLE','BUSINESS'])]

# Examine the price trend by plotting Price vs Category
ax = sns.stripplot(x = popular_app_cats["Updated_Price"], y = popular_app_cats["Category"], jitter=True, linewidth=1)
ax.set_title('App pricing trend across categories')

# Apps whose Price is greater than 200
apps_above_200 = popular_app_cats[['Category', 'App', 'Price']][popular_app_cats["Updated_Price"] > 200]
apps_above_200
```

**Filter out "junk" apps**

```python
# Select apps priced below $100
apps_under_100 = popular_app_cats[popular_app_cats["Updated_Price"]<100]

fig, ax = plt.subplots()
fig.set_size_inches(15, 8)

# Examine price vs category with the authentic apps
ax = sns.stripplot(x=apps_under_100["Updated_Price"], y=apps_under_100["Category"], data=apps_under_100,
                   jitter=True, linewidth=1)
ax.set_title('App pricing trend across categories after filtering for junk apps')
```

```python
#Task 9 : number of downloads of free vs paid apps
import plotly
import plotly.offline as py
import plotly.express as px
import plotly.graph_objs as go


trace0 = go.Box(
    # Data for paid apps
    y = df[df['Type'] == 'Paid']['Installs'],
    name = 'Paid'
)

trace1 = go.Box(
    # Data for free apps
    y = df[df['Type'] == 'Free']['Installs'],
    name = 'Free'
)

layout = go.Layout(
    title = "Number of downloads of paid apps vs. free apps",
    yaxis = dict(title = "Number of downloads",
                type = 'log',
                autorange = True)
)

# Add trace0 and trace1 to a list for plotting
data = [trace0, trace1]
plotly.offline.iplot({'data': data, 'layout': layout})
```
